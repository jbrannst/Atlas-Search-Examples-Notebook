# MongoDB Atlas Search: Dynamic paths

## Setup

```python
# import our libraries, instantiate our classes

import sys

from pymongo import MongoClient
import pprint
from datetime import datetime, tzinfo, timezone

client = MongoClient('mongodb+srv://main_user:8BjRQxoteAI6jFuB@cluster0.em4dc.mongodb.net')['nets']

pp = pprint.PrettyPrinter(indent=1)
```

```python
wildcard_search = [
    {
        '$search': {
            'index': 'default', 
            'compound': {
                'must': [
                    {
                        'range': {
                            'path': 'createdAt', 
                            'gt': datetime(2022, 12, 4, 0, 0, 0, tzinfo=timezone.utc), 
                            'lt': datetime(2022, 12, 4, 1, 0, 0, tzinfo=timezone.utc)
                        }
                    }
                ], 
                'should': [
                    {
                        'text': {
                            'path': {
                                'wildcard': 'events.*'
                            }, 
                            'query': 'Test product 001'
                        }
                    }
                ]
            }, 
            'highlight': {
                'path': {
                    'wildcard': 'events.*'
                }
            }
        }
    }, {
        '$project': {
            'merchantId': 1, 
            'paymentId': 1, 
            'createdAt': 1, 
            'updatedAt': 1, 
            'searchResults': {
                '$meta': 'searchHighlights'
            }
        }
    }, {
        '$unwind': {
            'path': '$searchResults'
        }
    }, {
        '$sort': {
            'searchResults.score': -1
        }
    }
]

result = list(client['payments'].aggregate(wildcard_search))
pp.pprint(result)
```
